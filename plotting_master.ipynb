```python
# From a directory
df = load_traces(
    directory=r"C:\Users\petermeisenheimer\Shepherd\data\PGRL418\D2\cd20\inverseshmoo_2us",
    pattern='3pp_*.csv',
    polarity_filter='npp',
    sort_by_polarity=True
)
start_ns, end_ns = 5e3,10e3
df = df.loc[(df.time_ns > start_ns) & (df.time_ns < end_ns)]

amplitudes = sorted(df['u_amplitude'].unique())
offset_scale = df['voltage_V'].std() *2 # Adjust multiplier to taste
offset_map = {amp: i * offset_scale for i, amp in enumerate(amplitudes)}
df['voltage_offset'] = df['voltage_V'] + df['u_amplitude'].map(offset_map)

fig, ax = plt.subplots(1, 1, figsize=(2, 2), dpi=200)
ax.set_facecolor('w')
m=sns.lineplot(data=df[::10],x='time_ns',y='voltage_offset',palette='managua',hue='u_amplitude')

# ax.set_xlim(5e3,10e3)
# ax.set_ylim(-.05,.05)
ax.legend(fontsize=6, title='previous pulse V', title_fontsize=6)
fig.savefig(r"C:\Users\petermeisenheimer\Shepherd\data\PGRL418\D2\cd20\inverseshmoo_2us\trace_2_npp.png")
```
